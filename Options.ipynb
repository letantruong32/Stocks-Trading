# FLOWALGO

### 1. Unusual

```python
df = pd.read_csv("option_export__04_20_2021.csv")
```

```python
premium = df['Premium'].str.replace('[$,#,@,&]', '').astype('int32')
df['Premium'] = premium
```

```python
unusual = df[df['Unusual'] == True]
```

```python
unusual.sort_values('Premium', ascending=False).head(60)
```

```python
unusual.to_csv('options/unusual_04_23.csv', index=False, header=True)
```
